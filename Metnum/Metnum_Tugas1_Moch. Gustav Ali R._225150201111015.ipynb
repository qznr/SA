```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bisection(fx):
    a = float(input('Masukkan nilai a : '))
    b = float(input('Masukkan nilai b : '))
    if (fx(a)*fx(b)) < 0:
        n = 0
        print('%8s\t%8s\t%8s\t%8s\t%8s\t%8s\t' % ('Iterasi','a', 'b', 'fx(a)', 'fx(b)', 'c'))
        while abs(fx(a)*fx(b)) > 0.000001:
            n += 1
            c = (a+b)/2
            print('%8d\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t' % (n,round(a, 3), round(b, 3), round(fx(a), 3), round(fx(b), 3), c))
            if fx(a) * fx(c) < 0.0:
                b = c
            else:
                a = c
        print('\nAkar persamaan adalah ', (c))
    else:
        print ('input a dan b invalid')
```

```python
def false_position(fx):
    a = float(input('Masukkan nilai a : '))
    b = float(input('Masukkan nilai b : '))
    cp = a
    Er = 10000.0
    if fx(a) * fx(b) < 0:
        n = 0
        print('%8s\t%8s\t%8s\t%8s\t%8s\t%8s\t' % ('Iterasi','a', 'b', 'c', 'fx(c)', 'Er'))
        while abs(Er) > 0.5:
            n += 1
            c = b - (fx(b) * (a-b))/(fx(a) - fx(b))
            cc = c
            Er = ((cc-cp)/cc)*100
            print('%8d\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t' % (n,a, b, c, fx(c), Er))
            cp = cc
            
            if fx(a) * fx(c) < 0.0:
                b = c
            else:
                a = c
        print('\nAkar persamaan adalah ', (c))
    else:
        print ('input a dan b invalid')
```

```python
x = np.arange(3,6,0.01)
fx = np.sin(10*x) + np.cos(3*x)
plt.figure()
plt.title('Grafik fungsi f(x)')
plt.xlabel('x')
plt.ylabel('f(x) = sin(10x) + cos (3x)')
plt.plot(x,fx,'-')
plt.grid(True)
plt.show()
```

Terlihat bahwa ada beberapa nilai yang melewati garis x, dengan kata lain yaitu memenuhi f(x) = 0.<br>
Saya memilih akar yang berada di interval 4.5 dan 5.0, maka dari itu saya menentukan untuk nilai a = 4.5 dan b = 5.0 

```python
def newton_method(fx, dfx):
    x0 = float(input('Masukkan nilai awal (x0) : '))
    Er = 10000.0
    n = 0

    print('%8s\t%8s\t%8s\t%8s\t%8s\t%8s\t' % ('Iterasi', 'x(k)', 'fx(k)', "f'x(k)", 'x(k+1)', 'Er'))

    while abs(Er) > 0.001:
        n += 1
        fx_val = fx(x0)
        dfx_val = dfx(x0)

        if dfx_val == 0:
            print('The derivative is zero. Newton method cannot continue.')
            return

        x1 = x0 - fx_val / dfx_val
        Er = ((x1 - x0) / x1) * 100
        print('%8d\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t' % (n, x0, fx_val, dfx_val, x1, Er))
        x0 = x1

    print('\nAkar persamaan adalah ', x0)


# Example usage of the newton_method function:
def f(x):
    return -(1 / np.sqrt(x**2 + 1)) + 1/3

def df(x):
    return (x*2 + 1)*(-1.5)

newton_method(f, df)
```

```python
def secant_method(fx, dfx):
    x0 = float(input('Masukkan nilai awal (x0) : '))
    x1 = float(input('Masukkan nilai awal (x1) : '))
    Er = 10000.0
    k = 0

    print('%8s\t%8s\t%8s\t%8s\t%8s\t%8s\t%8s\t' % ('k', 'x(k-1)', 'x(k)', 'fx(k)', "f'x(k)", 'x(k+1)', 'Er'))

    while abs(Er) > 0.00001:
        k += 1
        fx0 = fx(x0)
        fx1 = fx(x1)
        dfx0 = dfx(x0)

        if dfx0 == 0:
            print('The derivative is zero. Secant method cannot continue.')
            return

        x2 = x1 - (fx1 * (x0 - x1)) / (fx0 - fx1)
        Er = ((x2 - x1) / x2) * 100
        print('%8d\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t%8.5f\t' % (k, x0, x1, fx1, dfx0, x2, Er))
        x0 = x1
        x1 = x2

    print('\nAkar persamaan adalah ', x1)


# Example usage of the secant_method function:
def f(x):
    return -(1 / np.sqrt(x**2 + 1)) + 1/3

def df(x):
    return (x*2 + 1)*(-1.5)

secant_method(f, df)
```
